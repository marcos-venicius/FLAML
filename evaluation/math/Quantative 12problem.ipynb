```python
import json
import os

# Create an empty list to store the JSON data
data_list = []

# Specify the directory you want to scan for JSON files
dir_path = 'problems'

# Get a list of all the JSON files in the directory
json_files = [pos_json for pos_json in os.listdir(dir_path) if pos_json.endswith('.json')]

# Iterate over each JSON file
for filename in json_files:
    with open(os.path.join(dir_path, filename)) as json_file:
        # Load the JSON data from the file
        data = json.load(json_file)

        # Append the data to the list
        data_list.append(data)
        print(data["problem"])
        print("Answer: ", data["correct_ans"])
        print("------------------------")

# (Note: When writing code, always print the results.)
```

# AgentChat

```python
from flaml import oai
import os
from flaml.autogen.agentchat import AssistantAgent, UserProxyAgent
config_list = [
    {
        'model': 'gpt-4',
        'api_key': open("key_openai.txt", "r").read().strip(),
    },
]

# create an AssistantAgent named "assistant"
assistant = AssistantAgent(
    "assistant",
    oai_config={
        "seed": 42,
        "config_list": config_list,
    }
)
# create a UserProxyAgent instance named "user"
user = UserProxyAgent(
    "user",
    human_input_mode="NEVER",
    max_consecutive_auto_reply=10,
    is_termination_msg=lambda x: x.get("content", "").rstrip().endswith("TERMINATE") or x.get("content", "").rstrip().endswith('"TERMINATE".'),
    code_execution_config={
        "work_dir": "coding",
        "use_docker": False,  # set to True or image name like "python:3" to use docker
    },
)



for p in data_list:
    assistant.reset()
    user.initiate_chat(assistant, message=p['problem'])
    print("********************************************")
    print("********************************************")
```

# LangChain


```python
from langchain.chat_models import ChatOpenAI
api_key = open("key_openai.txt", "r").read().strip()
llm = ChatOpenAI(model_name='gpt-4', openai_api_key=api_key)

from langchain.agents import initialize_agent
from langchain.agents import AgentType
from langchain.tools.python.tool import PythonREPLTool

tools = [PythonREPLTool()]
agent = initialize_agent(tools, llm, agent=AgentType.ZERO_SHOT_REACT_DESCRIPTION, verbose=True)
```

```python
for p in data_list:
    result = agent.run(p['problem'])
    print("********************************************")
    print("********************************************")
```
